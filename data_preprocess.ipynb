## energy

```python
tempfile = ['Downloads/tepco_2016.txt', 'Downloads/tepco_2017.txt', 'Downloads/tepco_2018.txt',
       'Downloads/tepco_2019.txt', 'Downloads/tepco_2020.txt', 'Downloads/tepco_2021.txt']
temp = pd.DataFrame(columns=['DATE', 'TIME', 'ENE'])

for file in tempfile:
    time = []
    tmp = pd.read_csv(file)
    
    for row in tmp['TIME']:
        if len(row) < 5:
            time.append('0'+row)
        else:
            time.append(row)
    
    tmp['TIME'] = time
    tmp["DATE"] = pd.to_datetime(tmp["DATE"], infer_datetime_format=True)
    tmp.rename(columns = {'実績(万kW)' : 'ENE'}, inplace = True)
    temp = temp.append(tmp)
temp
```

```python
clmfile = ['Downloads/weather_tokyo_2016.csv', 'Downloads/weather_tokyo_2017.csv', 'Downloads/weather_tokyo_2018.csv',
          'Downloads/weather_tokyo_2019.csv', 'Downloads/weather_tokyo_2020.csv', 'Downloads/weather_tokyo_2021.csv']
clmt = pd.DataFrame(columns=['DATE', 'TIME', 'TEMP', 'FEEL'])

for file in clmfile:
    df = pd.read_csv(file)
    
    idxs = []
    tmp = 0
    
    for row in df['datetime']:
        if row.split(' ')[1][3:5] != '00':
            idxs.append(tmp)
        tmp += 1
    
    df.drop(idxs, inplace=True)
    
    date = []
    time = []
    
    for row in df['datetime']:
        date.append(row.split(' ')[0])
        time.append(row.split(' ')[1][:5])

    
    
    
    
    df['DATE'] = date
    df['TIME'] = time
    df = df[['DATE', 'TIME', 'temp(F)', 'feels_like(F)']]
    df['DATE'] = pd.to_datetime(df['DATE'], infer_datetime_format=True)
    df.rename(columns = {'temp(F)' : 'TEMP'}, inplace = True)
    df.rename(columns = {'feels_like(F)' : 'FEEL'}, inplace = True)
    
    clmt = clmt.append(df)
    
clmt.head()
```
